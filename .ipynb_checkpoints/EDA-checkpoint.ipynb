```python
ahead_elec = pd.read_csv(os.path.join('./data', '제주전력시장_시장전기가격_하루전가격.csv'))
rtime_elec = pd.read_csv(os.path.join('./data', '제주전력시장_시장전기가격_실시간가격.csv'))
```

```python
plt.figure(figsize=(20, 5))
plt.plot(ahead_elec['하루전가격(원/kWh)'][:500], label = "ahead prices")
plt.plot(rtime_elec['실시간 임시 가격(원/kWh)'][:500], label = 'Real-time temporary prices')
plt.plot(rtime_elec['실시간 확정 가격(원/kWh)'][:500], label = 'Real-time confirmed price')
plt.legend()
plt.show()
```
