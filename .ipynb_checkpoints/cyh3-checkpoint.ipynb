```python
import pandas as pd

weather_actual_1 = pd.read_csv("기상실측데이터_1.csv", encoding="utf-8")
weather_actual_2 = pd.read_csv("기상실측데이터_2.csv", encoding="utf-8")
weather_forecast_1 = pd.read_csv("기상예측데이터_1.csv", encoding="utf-8")
weather_forecast_2 = pd.read_csv("기상예측데이터_2.csv", encoding="utf-8")
price_real_time = pd.read_csv('제주전력시장_시장전기가격_실시간가격.csv')
price_day_ahead = pd.read_csv('제주전력시장_시장전기가격_하루전가격.csv')
market_status = pd.read_csv('제주전력시장_현황데이터.csv')
```

```python
# 'ts' 컬럼을 datetime으로 변환
weather_actual_1['datetime'] = pd.to_datetime(weather_actual_1['ts'], errors="coerce",unit='s')
weather_actual_2['datetime'] = pd.to_datetime(weather_actual_2['ts'], errors="coerce",unit='s')
weather_forecast_1['datetime'] = pd.to_datetime(weather_forecast_1['ts'], errors="coerce",unit='s')
weather_forecast_2['datetime'] = pd.to_datetime(weather_forecast_2['ts'], errors="coerce",unit='s')
price_real_time['datetime'] = pd.to_datetime(price_real_time['ts'], errors="coerce",unit='s')
price_day_ahead['datetime'] = pd.to_datetime(price_day_ahead['ts'], errors="coerce",unit='s')
market_status['datetime'] = pd.to_datetime(market_status['ts'], errors="coerce",unit='s')

# 필요에 따라 기존 'ts' 컬럼 삭제
weather_actual_1.drop(columns=['ts'], inplace=True)
weather_actual_2.drop(columns=['ts'], inplace=True)
weather_forecast_1.drop(columns=['ts'], inplace=True)
weather_forecast_2.drop(columns=['ts'], inplace=True)
price_real_time.drop(columns=['ts'], inplace=True)
price_day_ahead.drop(columns=['ts'], inplace=True)
market_status.drop(columns=['ts'], inplace=True)

```

```python
for df in [weather_actual_1, weather_actual_2, weather_forecast_1, weather_forecast_2, price_real_time, price_day_ahead, market_status]:
    df['year'] = df['datetime'].dt.year
    df['month'] = df['datetime'].dt.month
    df['day'] = df['datetime'].dt.day
    df['hour'] = df['datetime'].dt.hour
    df['day_of_week'] = df['datetime'].dt.dayofweek  # 0=월요일, 6=일요일
    df['is_weekend'] = df['day_of_week'] >= 5  # 주말 여부
# weather_actual_1.drop(columns=['datetime'], inplace=True)
# weather_actual_2.drop(columns=['datetime'], inplace=True)
# weather_forecast_1.drop(columns=['datetime'], inplace=True)
# weather_forecast_2.drop(columns=['datetime'], inplace=True)
# price_real_time.drop(columns=['datetime'], inplace=True)
# price_day_ahead.drop(columns=['datetime'], inplace=True)
# market_status.drop(columns=['datetime'], inplace=True)
```

```python
# 2024년 3월 1일 이후 데이터만 남기기
start_date = pd.to_datetime("2024-03-01")

weather_actual_1 = weather_actual_1[weather_actual_1["datetime"] >= start_date]
weather_actual_2 = weather_actual_2[weather_actual_2["datetime"] >= start_date]
weather_forecast_1 = weather_forecast_1[weather_forecast_1["datetime"] >= start_date]
weather_forecast_2 = weather_forecast_2[weather_forecast_2["datetime"] >= start_date]
price_real_time = price_real_time[price_real_time["datetime"] >= start_date]
price_day_ahead = price_day_ahead[price_day_ahead["datetime"] >= start_date]
market_status = market_status[market_status["datetime"] >= start_date]
```

```python
import pandas as pd

# 'datetime' 컬럼을 1시간 단위로 라운딩
weather_actual_1['datetime'] = pd.to_datetime(weather_actual_1['datetime']).dt.round('H')
weather_actual_2['datetime'] = pd.to_datetime(weather_actual_2['datetime']).dt.round('H')
weather_forecast_1['datetime'] = pd.to_datetime(weather_forecast_1['datetime']).dt.round('H')
weather_forecast_2['datetime'] = pd.to_datetime(weather_forecast_2['datetime']).dt.round('H')
price_real_time['datetime'] = pd.to_datetime(price_real_time['datetime']).dt.round('H')
price_day_ahead['datetime'] = pd.to_datetime(price_day_ahead['datetime']).dt.round('H')
market_status['datetime'] = pd.to_datetime(market_status['datetime']).dt.round('H')

# 1. 기상 실측 및 예측 데이터 병합 (1시간 단위로 라운딩 후 병합)
weather_actual = pd.concat([weather_actual_1, weather_actual_2], ignore_index=True)
weather_forecast = pd.concat([weather_forecast_1, weather_forecast_2], ignore_index=True)
weather_data = weather_actual.merge(weather_forecast, on=['datetime', 'location'], suffixes=('_actual', '_forecast'), how='outer')

# 2. 전력 가격 데이터와 시장 상태 데이터 병합 (1시간 단위로 라운딩 후 병합)
data = weather_data.merge(price_real_time, on='datetime', how='outer')
data = data.merge(price_day_ahead, on='datetime', how='outer')
data = data.merge(market_status, on='datetime', how='outer')

# location 컬럼 제거 (필요하지 않다면)
data.drop(columns=['location'], inplace=True)

# 3. 결측치 처리 - 남은 결측치를 평균으로 채움
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_cols] = data[numeric_cols].fillna(data[numeric_cols].mean())

# 최종 데이터 확인
print(data.head())
```

```python
# 앞선 값으로 채우기 (time series에서 연속성을 고려할 때 유용)
data.fillna(method='ffill', inplace=True)

# 혹은 뒤의 값으로 채우기
data.fillna(method='bfill', inplace=True)
```

```python
data
```

```python
# 사용할 주요 feature 리스트
features = [
    'temp_actual', 'rel_hum_actual', 'wind_spd_actual',
    'temp_forecast', 'rel_hum_forecast', 'wind_spd_forecast',
    '실시간 확정 가격(원/kWh)', '공급능력(kW)', '현재 수요(kW)',
    '태양광 발전량kW)', '풍력 발전량(kW)', '공급 예비력(kW)',
    'hour', 'day_of_week', 'is_weekend'
]
```

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 데이터 정렬 (datetime을 기준으로)
data = data.sort_values(by='datetime')

# 80%까지는 훈련 데이터로, 20%는 테스트 데이터로 사용
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Feature와 Target 설정
X_train, y_train = train[features], train['하루전가격(원/kWh)']
X_test, y_test = test[features], test['하루전가격(원/kWh)']

# 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 모델 학습
model = LGBMRegressor(random_state=42)
model.fit(X_train_scaled, y_train)

# 예측
y_pred = model.predict(X_test_scaled)

# 성능 평가
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
import matplotlib.pyplot as plt

# 실제값 vs 예측값 비교 그래프
plt.figure(figsize=(14, 6))
plt.plot(y_test.values, label="Actual Values", color="blue", alpha=0.6)
plt.plot(y_pred, label="Predicted Values", color="red", alpha=0.6)
plt.title("Actual vs Predicted Electricity Prices (Time-based Split)")
plt.xlabel("Time")
plt.ylabel("Electricity Price (원/kWh)")
plt.legend()
plt.show()

# 오차 분포 히스토그램
errors = y_test.values - y_pred
plt.figure(figsize=(8, 6))
plt.hist(errors, bins=30, color="purple", alpha=0.7)
plt.title("Distribution of Prediction Errors")
plt.xlabel("Error (Actual - Predicted)")
plt.ylabel("Frequency")
plt.show()
```

```python
feature_importances = pd.DataFrame({
    'feature': X_train.columns,
    'importance': model.feature_importances_
})
feature_importances = feature_importances.sort_values(by='importance', ascending=False)
print(feature_importances)
```

```python
threshold = 10
selected_features = feature_importances[feature_importances['importance'] > threshold]['feature'].tolist()

# 선택된 특징만 사용
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression  # RFE에 사용할 기본 모델
from sklearn.preprocessing import StandardScaler

# 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# RFE 초기화 (LightGBM이 아닌 간단한 선형 회귀 모델로 선택)
selector = RFE(estimator=LinearRegression(), n_features_to_select=10, step=1)
selector = selector.fit(X_train_scaled, y_train)

# 선택된 특징만 선택
selected_features = X_train.columns[selector.support_]
X_train_selected = X_train[selected_features]
X_test_selected = X_test[selected_features]

print("Selected Features:", selected_features)
```

```python
# 하루 전 가격과 7일 전 가격을 새로운 특징으로 추가
data['price_lag_1'] = data['하루전가격(원/kWh)'].shift(100)
data['price_lag_7'] = data['하루전가격(원/kWh)'].shift(24)

# 결측치 처리 (초기 몇 개 행이 결측치가 될 수 있음)
data.fillna(method='bfill', inplace=True)  # 뒤의 값으로 채우기

# feature 리스트에 새로 생성한 lag 변수 추가
features = features + ['price_lag_1', 'price_lag_7']
```

```python
X = data[features]
X
```

```python

# 2. Feature와 Target 설정
X = data[features]
y = data['하루전가격(원/kWh)']  # 타겟 변수 설정

# 3. 학습 및 검증 데이터 분할 (Time-based Split)
# 시계열 데이터이므로, 과거 데이터를 학습 데이터로, 이후 데이터를 검증 데이터로 사용
train_size = int(len(data) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 4. 데이터 스케일링
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 5. 모델 학습
model = LGBMRegressor(random_state=42)
model.fit(X_train_scaled, y_train)

# 6. 예측 및 평가
y_pred = model.predict(X_test_scaled)

# 성능 평가
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
