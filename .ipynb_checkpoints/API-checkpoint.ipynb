```python
from datetime import datetime, timezone, timedelta

# 유닉스 타임스탬프 값
ts = 1709222220

# UTC 시간을 KST로 변환
def unix_to_kst(timestamp):
    # 타임스탬프를 datetime으로 변환하고 9시간을 더해 KST로 조정
    dt = datetime.fromtimestamp(timestamp, tz=timezone.utc) + timedelta(hours=9)
    return dt.strftime('%Y-%m-%d %H:%M:%S')

# 결과 출력
print("ts (KST):", unix_to_kst(ts))
```
