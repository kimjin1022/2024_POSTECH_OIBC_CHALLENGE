```python
import pandas as pd

base_path = "/home/jinjinjara1022/OBIC/data/"

# 데이터를 불러오는 코드
available_data = {
    #"기상실측데이터_1": pd.read_csv(base_path + "기상실측데이터_1.csv", encoding="utf-8"),
    #"기상실측데이터_2": pd.read_csv(base_path + "기상실측데이터_2.csv", encoding="utf-8"),
    "기상예측데이터_1": pd.read_csv(base_path + "기상예측데이터_1.csv", encoding="utf-8"),
    "기상예측데이터_2": pd.read_csv(base_path + "기상예측데이터_2.csv", encoding="utf-8"),
    "하루전가격데이터": pd.read_csv(base_path + "제주전력시장_시장전기가격_하루전가격.csv", encoding="utf-8"),
    #"현황데이터": pd.read_csv(base_path + "제주전력시장_현황데이터.csv", encoding="utf-8")
}
```

```python
# 모든 데이터프레임에서 'ts' 열을 int64로 변환
for key, df in available_data.items():
    if 'ts' in df.columns:
        df['ts'] = pd.to_numeric(df['ts'], errors='coerce')
        df.dropna(subset=['ts'], inplace=True)
        df['ts'] = df['ts'].astype(int)
```

```python
# 하루전가격데이터를 기준으로 병합 시작
merged_data = available_data['하루전가격데이터']

# 가장 가까운 시간으로 병합, 각 데이터에 접미사 추가
for key, df in available_data.items():
    if key != '하루전가격데이터':
        suffix = f"_{key}"  # 접미사 추가
        merged_data = pd.merge_asof(
            merged_data.sort_values('ts'),      # 기준 데이터 정렬
            df.sort_values('ts').add_suffix(suffix),  # 접미사 추가 후 정렬
            left_on='ts', right_on=f'ts{suffix}',     # 기준 컬럼 일치
            direction='nearest'                  # 가장 가까운 시간 기준으로 병합
        )
```

```python
data = merged_data
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# 숫자형 열만 선택하여 정규화
numeric_features = data.select_dtypes(include=[np.number]).columns.difference(['하루전가격(원/kWh)'])  # 타겟 변수 제외
scaler = MinMaxScaler()
data[numeric_features] = scaler.fit_transform(data[numeric_features])

# 3. 특성 생성
# (1) 요일 변수 생성 (ts가 Unix timestamp일 경우)
data['date'] = pd.to_datetime(data['ts'], unit='s')
data['weekday'] = data['date'].dt.weekday  # 0=월요일, 6=일요일
data['hour'] = data['date'].dt.hour

# (2) 이동 평균 추가 (전기 가격의 3시간 이동 평균 예시)
#data['하루전가격_3시간_이동평균'] = data['하루전가격(원/kWh)'].rolling(window=3, min_periods=1).mean()

# 불필요한 열 제거
data.drop(columns=['date', 'ts'], inplace=True)  # date와 ts는 제거해도 됨

# 4. 타겟 변수와 입력 변수 분리
X = data.drop(columns=['하루전가격(원/kWh)'])  # 입력 변수
y = data['하루전가격(원/kWh)']                 # 타겟 변수
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ts 데이터를 설정 (예: data['하루전가격(원/kWh)'])
ts = data['하루전가격(원/kWh)']

# 데이터 스케일링 (MinMaxScaler 사용)
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(ts.values.reshape(-1, 1))

# 시계열 데이터셋 생성 (입력 시퀀스 길이: timesteps)
timesteps = 48
X, y = [], []
for i in range(len(data_scaled) - timesteps):
    X.append(data_scaled[i:i+timesteps].flatten())  # 과거 48개의 데이터를 하나의 배열로
    y.append(data_scaled[i+timesteps])              # 예측할 대상 값

X, y = np.array(X), np.array(y)
```

```python
# 훈련, 테스트 데이터 분할
from sktime.forecasting.model_selection import temporal_train_test_split
X_train, X_test, y_train, y_test = temporal_train_test_split(X, y, test_size=0.2)

# LGBMRegressor 모델 설정 및 학습
model = lgb.LGBMRegressor(
    objective='regression',
    boosting_type='gbdt',
    learning_rate=0.1,
    num_leaves=31,
    max_depth=-1,
    min_data_in_leaf=20,
    n_estimators=500  # 충분히 큰 n_estimators 설정
)

# 모델 학습
model.fit(X_train, y_train)
```

```python
# 예측
predicted = model.predict(X_test)
predicted = scaler.inverse_transform(predicted.reshape(-1, 1))
y_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# 평가 지표 계산
mse = mean_squared_error(y_actual, predicted)
mae = mean_absolute_error(y_actual, predicted)
r2 = r2_score(y_actual, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R^2 Score: {r2}")

# 예측 결과 시각화 (마지막 200개 데이터만 표시)
visualization_range = 1000
y_actual_partial = y_actual[-visualization_range:]
predicted_partial = predicted[-visualization_range:]

plt.figure(figsize=(50, 5))
plt.plot(y_actual_partial, label="Actual Price", color="blue", linestyle="-", linewidth=2)
plt.plot(predicted_partial, label="Predicted Price", color="red", linestyle="--", linewidth=2)
plt.xlabel("Time")
plt.ylabel("Electricity Price")
plt.legend(loc="upper right")
plt.title("Actual vs Predicted Electricity Price (LightGBM Model)")
plt.grid(True)
plt.show()
```

```python
# Define SMAPE loss function
def SMAPE(true, pred):
    return np.mean((np.abs(true-pred))/(np.abs(true) + np.abs(pred))) * 100
```
